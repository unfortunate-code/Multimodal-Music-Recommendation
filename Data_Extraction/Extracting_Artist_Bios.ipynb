### MSD - A Dataset

```python
import tarfile
import re
import pandas as pd

bios_dict = {'Artist ID':[], 'Bio':[]}
tar = tarfile.open("bios_msd-a.tar.gz", "r:gz")
for member in tar.getmembers():
    f = tar.extractfile(member)
    if f is not None:
        content = f.read()
        bios_dict['Artist ID'].append(re.split('\.|/', member.path)[1])
        bios_dict['Bio'].append(content)
```

```python
artist_bios = pd.DataFrame.from_dict(bios_dict)
```

```python
artist_bios = artist_bios.set_index('Artist ID')
```

### Wikipedia Scrapping

```python
import requests
import time
from bs4 import BeautifulSoup
import re

def get_bio(artist_name):
    response = requests.get("https://en.wikipedia.org/wiki/"+artist_name)
    if response.status_code != 200:
        print(artist_name)
        return ""
    soup = BeautifulSoup(response.content, 'html.parser')
    try:
        para = soup.find_all('p')
        for p in para:
            bio = p.get_text()
            if re.search('[a-zA-Z]', bio):
                bio = re.sub('\[\d+\]', '', bio.strip())
                break
        return bio
    except:
        print(artist_name)
        return ""
```
