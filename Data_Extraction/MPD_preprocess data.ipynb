```python
import json
import glob
from tqdm import tqdm
```

```python
tot_song_set = set()
song2idx = dict()
idx2song = dict()
offset = 0
playlists = []
song_counter = dict()
```

```python
def extract_playlist(playlist):
    global offset, song_counter
    pl = []
    for song in playlist:
        song_uri = song['track_uri']
        song_id = song2idx.get(song_uri, offset)
        if song_uri in song2idx:
            song_idx = song2idx[song_uri]
        else:
            song2idx[song_uri] = offset
            idx2song[offset] = song_uri
            song_idx = offset
            song_counter[song_idx] = 0
            offset+=1
        pl.append(song_idx)
        tot_song_set.add(song['track_uri'])
        song_counter[song_idx] += 1
        
    return pl
```

```python
for name in tqdm(glob.glob('data/*.json')):
    with open(name, 'r') as f:
        data = f.read()
        data = json.loads(data)
        data = data['playlists']
        for playlist in data:
            playlists.append(extract_playlist(playlist["tracks"]))
```

```python
len(song_counter)
```

```python
freq_song_list = []
for k, v in song_counter.items():
    if v >= 100:
        freq_song_list.append(k)
```

```python
len(freq_song_list)
```

```python
filtered_playlists = []
for pl in tqdm(playlists):
    pl_filtered = [i for i in pl if i in freq_song_list]
    if len(pl_filtered) > 3:
        filtered_playlists.append(pl_filtered)
```

```python
new_song_id2old_song_id = dict()
old_song_id2new_song_id = dict()
new_offset = 0
```

```python
filtered_playlists_new_id = []
for pl in tqdm(filtered_playlists):
    new_pl = []
    for song in pl:
        if song in old_song_id2new_song_id:
            new_pl.append(old_song_id2new_song_id[song])
        else:
            old_song_id2new_song_id[song] = new_offset
            new_song_id2old_song_id[new_offset] = song
            new_pl.append(old_song_id2new_song_id[song])
            new_offset += 1
        
    filtered_playlists_new_id.append(new_pl)
    
            
        
```

```python
import pickle
with open('output/playlist.data', 'wb') as f:
    pickle.dump(filtered_playlists_new_id, f)
```

```python
song2new_song_id = dict()
new_song_id2song = dict()
for old_song_id in old_song_id2new_song_id:
    song = idx2song[old_song_id]
    new_song_id = old_song_id2new_song_id[old_song_id]
    
    song2new_song_id[song] = new_song_id
    new_song_id2song[new_song_id] = song
```

```python
with open('output/idx2song.data', 'wb') as f:
    pickle.dump(new_song_id2song, f)
```

```python
song_id2info = dict()
artist_uri2artist_name = dict()
for name in tqdm(glob.glob('data/*.json')):
    with open(name, 'r') as f:
        data = f.read()
        data = json.loads(data)
        data = data['playlists']
        for playlist in data:
            for song in playlist["tracks"]:
                if song['track_uri'] in song2new_song_id:
                    song_dict = dict()
                    song_dict['artist_name'] = song['artist_name']
                    song_dict['artist_uri'] = song['artist_uri']
                    song_dict['track_name'] = song['track_name']
                    song_id2info[song['track_uri']] = song_dict
                    
                    artist_uri2artist_name[song['artist_uri']] = song['artist_name']
                
```

```python
with open('output/artist.data', 'wb') as f:
    pickle.dump(artist_uri2artist_name, f)
```

```python
with open('output/song_id2info.data', 'wb') as f:
    pickle.dump(song_id2info, f)
```
