```python
from os import listdir
from os.path import isfile, join
albums = [join('../Data/images', f) for f in listdir('../Data/images') if isfile(join('../Data/images', f)) and f.startswith('album')]
```

```python
from PIL import Image
index = 0
for album in albums:
    image = Image.open(album)
    width, height = image.size
    print(width, height)
    image.close()
    index += 1
    if index > 100:
        break
```
